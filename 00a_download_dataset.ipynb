```python
import os
BASE = '/content/drive/MyDrive/YeastCaduceus/data/backup'

folders = [
    # Unsupervised — raw genomes
    'genomes/R64',
    'genomes/80_strains',
    'genomes/165_Saccharomycetales',
    'genomes/1341_Fungus',
    # Unsupervised — processed TFRecords
    'tfrecords/R64',
    'tfrecords/80_strains',
    'tfrecords/165_Saccharomycetales',
    'tfrecords/1341_Fungus',
    # Supervised
    'supervised/bigwigs',
    'supervised/processed',
]

for f in folders:
    os.makedirs(f'{BASE}/{f}', exist_ok=True)
print('✅ Folder structure created')
```

```python
buckets = {
    'genomes/R64':                     'gs://shorkie-paper/data/unsupervised/genome/R64/',
    'genomes/80_strains':              'gs://shorkie-paper/data/unsupervised/genome/80_strains/',
    'genomes/165_Saccharomycetales':   'gs://shorkie-paper/data/unsupervised/genome/165_Saccharomycetales/',
    'genomes/1341_Fungus':             'gs://shorkie-paper/data/unsupervised/genome/1341_Fungus/',
    'tfrecords/R64':                   'gs://shorkie-paper/data/unsupervised/processed/R64/',
    'tfrecords/80_strains':            'gs://shorkie-paper/data/unsupervised/processed/80_strains/',
    'tfrecords/165_Saccharomycetales': 'gs://shorkie-paper/data/unsupervised/processed/165_Saccharomycetales/',
    'tfrecords/1341_Fungus':           'gs://shorkie-paper/data/unsupervised/processed/1341_Fungus/',
    'supervised/bigwigs':              'gs://shorkie-paper/data/supervised/bigwigs/',
    'supervised/processed':            'gs://shorkie-paper/data/supervised/processed/',
}

for label, uri in buckets.items():
    print(f'\n--- {label} ---')
    !gsutil -u {GCP_PROJECT} du -sh {uri}
```

```python
for label, uri in buckets.items():
    dest = f'{BASE}/{label}/'
    print(f'\n⬇️  Downloading {label} ...')
    !gsutil -u {GCP_PROJECT} -m cp -r {uri}* {dest}
    print(f'✅ {label} done')

# --- Cell 6: Verify ---
for label in buckets:
    path = f'{BASE}/{label}'
    count = sum(len(f) for _, _, f in os.walk(path))
    print(f'{label}: {count} files')

print('\nTotal size:')
!du -sh {BASE}
```
